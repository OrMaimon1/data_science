```python
#imports mec-learning

from sklearn.preprocessing import LabelEncoder, LabelBinarizer
import numpy as np
import pandas as pd
import sklearn
from scipy import stats
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.preprocessing import MinMaxScaler, StandardScaler, normalize
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, f1_score
from sklearn import linear_model
from sklearn import metrics,preprocessing
from matplotlib import pyplot as plt
%matplotlib inline 
```

```python
clean_data = pd.read_csv("GameData.csv",index_col=0)
```

```python
df_m =pd.read_csv("GameData.csv",index_col=0)
pd.to_datetime(df_m.Release_Date)
df_m['Release_Date_Year'] = pd.to_datetime(clean_data['Release_Date']).dt.year
df_m.drop(["Last_Update","Release_Date"],axis =1 ,inplace = True)
df_m["Game"]=LabelEncoder().fit_transform(df_m["Game"])
df_m["Console"]=LabelEncoder().fit_transform(df_m["Console"])
df_m["Publisher"]=LabelEncoder().fit_transform(df_m["Publisher"])
df_m["hasScoreBiggerThan5"]=df_m["Critic_Score"]>7
df_m.drop(["Critic_Score"],axis =1,inplace = True)

df_m["hasScoreBiggerThan5"] = df_m["hasScoreBiggerThan5"].astype(int)


xtrain,xtest,ytrain,ytest =train_test_split(df_m.drop("hasScoreBiggerThan5",axis=1),df_m["hasScoreBiggerThan5"],random_state=0)
clf =LogisticRegression()
clf.fit(xtrain,ytrain)
acc = clf.score(xtest,ytest)
print(f"acc is : {acc}")

```
